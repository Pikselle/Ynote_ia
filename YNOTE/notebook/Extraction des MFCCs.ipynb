```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
theo1 = 'sample/Bobby McFerrin - Dont Worry Be Happy (Official Music Video).mp3'
theo2 = 'sample/Rick Astley - Never Gonna Give You Up (Official Music Video).mp3'
```

```python
def calculer_mfcc(filename):
    y, sr = librosa.load(filename, sr=None)
    # Calcul des MFCC
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
    return y, sr, mfcc
```

```python
# Charger l'audio
filename = test

y, sr, mfcc = calculer_mfcc(filename)

# Afficher les MFCC
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfcc, x_axis='time', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title("MFCC (Mel-Frequency Cepstral Coefficients)")
plt.xlabel("Temps")
plt.ylabel("Coefficients MFCC")
plt.show()
```

```python
# Charger les deux fichiers audio
def load_audio(file):
    y, sr = librosa.load(file, sr=None)
    return y, sr

# Calculer les MFCC
def compute_mfcc(y, sr, n_mfcc=20):
    return librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)

# Afficher les MFCC
def plot_mfcc(mfcc, sr, title):
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfcc, x_axis='time', sr=sr)
    plt.colorbar()
    plt.title(title)
    plt.show()

# Fichiers audio
file1 = theo1
file2 = theo2

# Charger les fichiers audio
y1, sr1 = load_audio(file1)
y2, sr2 = load_audio(file2)

# Calculer les MFCC
mfcc1 = compute_mfcc(y1, sr1)
mfcc2 = compute_mfcc(y2, sr2)

# Visualiser les MFCC
plot_mfcc(mfcc1, sr1, "MFCC de la musique 1")
plot_mfcc(mfcc2, sr2, "MFCC de la musique 2")
```

Les MFCC (Mel-Frequency Cepstral Coefficients) sont des caractéristiques couramment utilisées en traitement du signal audio, notamment pour la reconnaissance vocale et l'analyse de sons. Voici une explication concise de ce que représentent les MFCC et ce qu'ils permettent de faire :

Conversion du signal en spectrogramme : Le signal audio est découpé en trames (frames) pour capturer l'évolution temporelle. Chaque trame est transformée en domaine fréquentiel via la Transformée de Fourier.

Application de l'échelle de Mel : L'échelle de Mel est une échelle perceptive qui reflète la façon dont l'oreille humaine perçoit les fréquences. Les fréquences plus basses sont échantillonnées plus précisément que les fréquences élevées.

Extraction des coefficients cepstraux : Une Transformée de Fourier inverse est appliquée pour obtenir une représentation compacte du spectre de puissance, appelée cepstre. Les coefficients cepstraux issus de cette représentation sont les MFCC.

Avantages des MFCC
Invariance aux changements de puissance : Ils modélisent les caractéristiques spectrales tout en réduisant la sensibilité aux variations d'intensité sonore.
Compacte : Généralement, seuls les premiers coefficients sont nécessaires pour une analyse efficace.
